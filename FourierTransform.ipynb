**Import Python modules**

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# create a black image with a white dot
img = np.zeros((256, 256), dtype=np.uint8)
img[128, 128] = 255

# Apply Fourier transform
dft = cv2.dft(np.float32(img), flags=cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)

# Calculate the magnitude and phase spectrum
magnitude_spectrum = 20 * np.log(cv2.magnitude(dft_shift[:, :, 0], dft_shift[:, :, 1]))
phase_spectrum = np.angle(dft_shift[:, :, 0] + 1j*dft_shift[:, :, 1])

# Plot the input image, magnitude and phase spectrum
fig, axs = plt.subplots(1, 3, figsize=(10, 10))
plt.subplot(131), plt.imshow(img, cmap='gray'), plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(magnitude_spectrum, cmap='gray'), plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(phase_spectrum, cmap='gray'), plt.title('Phase Spectrum'), plt.xticks([]), plt.yticks([])
plt.tight_layout(pad=1)
plt.show()
```

In this example, we first create a binary image with a white square on a black background using the **np.zeros()** and **cv2.rectangle()** functions. Then we perform Fourier transform using the **cv2.dft()** function and shift the zero-frequency component to the center of the image using the **np.fft.fftshift()** function. We then calculate the magnitude spectrum of the Fourier transform using the **cv2.magnitude()** function and convert it to a logarithmic scale using the **np.log()** function. Finally, we display the input image and magnitude spectrum.

Note that the magnitude_spectrum matrix contains complex values. You can use **np.real()** and **np.imag()** functions to extract the real and imaginary parts of the Fourier transform respectively.

**Fourier Transform of a 2D Sine Wave** 

```python
# Create a horizontal 2D sine wave image
wavelength = 5
amplitude = 100
img = np.zeros((512, 512), dtype=np.uint8)
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        img[i,j] = amplitude * np.sin(i/wavelength)

# Apply Fourier transform
dft = cv2.dft(np.float32(img), flags=cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)

# Calculate the magnitude and phase spectrum
magnitude_spectrum = 20 * np.log(cv2.magnitude(dft_shift[:, :, 0], dft_shift[:, :, 1]))
phase_spectrum = np.angle(dft_shift[:, :, 0] + 1j*dft_shift[:, :, 1])

# Plot the input image, magnitude and phase spectrum
fig, axs = plt.subplots(1, 3, figsize=(10, 10))
plt.subplot(131), plt.imshow(img, cmap='gray'), plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(magnitude_spectrum, cmap='gray'), plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(phase_spectrum, cmap='gray'), plt.title('Phase Spectrum'), plt.xticks([]), plt.yticks([])
plt.tight_layout(pad=1)
plt.show()
```
